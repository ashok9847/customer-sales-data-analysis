```python
df.dropna(subset=['Customer_ID'], inplace=True)
# find any rows where the ID is missing (NaN or null), and permanently delete those entire rows from your DataFrame
```

```python
df['Age'].unique()
```

```python
# int(df['Age'])
# df['Age']
import re
def extract(age):
  age_num=re.findall('[0-9]+',str(age))
  if len(age_num)>0:
    return age_num[0]
  else:
    return age

df['Age']=df['Age'].apply(lambda x:extract(x))
df['Age']
```

```python

# df_age=df['Age'].isnull()
# df_age
# df.isnull().sum()
# df_age=df[df['Age'].notnull()]['Age']
# df_age = df[df['Age'] != 'nan']['Age']
# df_age.isnull().sum()

# df_age.isnull().sum()
# age_median=int(df['Age'].dropna().astype('int64').median())
# age_median
df['Age'].unique()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')#converted every string into num and if we encounter with any value with is not num then it will return NAN
#coz we alread have nan in my table so it will return flaot
df['Age'].unique()
age_median=int(df['Age'].dropna().astype('int64').median())
#here we have to cualte the median() without the null values we do dropna

df['Age'] = df['Age'].fillna(age_median)
#fill the nan with the median
df['Age'] = df['Age'].astype('int64')
# change the str type into num now
df['Age'].unique()
df['Age'].dtype
```

```python
price_median=int(df['Purchase_Amount'].median())
price_median
```

```python
df['Purchase_Amount']=df['Purchase_Amount'].fillna(price_median)
df['Purchase_Amount'] = df['Purchase_Amount'].astype('int64')

df['Purchase_Amount'].unique()
```

```python
df.isnull().sum()
```

```python
for col in ['Gender','City','Feedback_Score','Country']:
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
for col in ['Last_Purchase_Date']:
  df[col] = df[col].fillna(method='ffill')
```
